    **O Teorema de bayes diz que :**
    
    
$$ P(C|+) = \frac{P(+|C) \ P(C)}{P(+|C) \ P(C)  + P(+|¬ C) \ P(¬ C)}    $$

```python
p_c = 0.6
p_posc = p_c * 0.96
p_posnc = p_c * 0.01
p_nc = 0.4
```

```python
# Pelo teorema de bayes, temos que:
p_cpos = (p_posc * p_c) / ((p_posc * p_c) + (p_posnc * p_nc))
p_cpos*100
```

```python
mu = 7
sigma = 2
x0 = 4
x1 = 11

z0 = (x0 - mu) / sigma
z1 = (x1 - mu) / sigma

p0 = round((st.norm.cdf(z0))*100, 2)
p1 = round((st.norm.cdf(z1))*100, 2)
p = round(p1 - p0,2)

print('Z0: ',z0)
print('Z1: ',z1)
print('=================================================')
print('Probabilidade Z0 = ', p0, '%')
print('Probabilidade Z1 = ', p1, '%')
print('=================================================')
print('Probabilidade Z0 > P < Z1 = ', p, '%')
```

## *H1 mu <= 35*

```python
mu = 35
n = 20
amostras = [37, 34, 36, 37, 35, 32, 30, 35, 34, 36, 39, 38, 34, 35, 31, 30, 33, 35, 36, 31]
xmed = np.mean(amostras)
variancia = 4
sigma = np.sqrt(variancia)
alfa = 0.05
```

```python
z = st.norm.ppf(alfa)

xc = round(z*(sigma/np.sqrt(n)) + mu,2)

print('Probabilidade para 95%:', z)
print('Valor crítico:', xc,'mg')
print('Valor observado:', xmed,'mg')
print("Teste de Hipótese: ")
if(xmed > xc):
    print("Aceitamos H0")
else:
    print("Rejeitamos H0")
```
